```python
import os
import pandas as pd
import numpy as np
import folium as f
import datetime 
from folium.plugins import HeatMap, HeatMapWithTime
```

Found on Shane Lynn Data Science blog https://www.shanelynn.ie/the-irish-property-price-register-geocoded-to-small-areas/

```python
df = pd.read_csv('ppr_data_encoded.csv', ',', header=0, encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
norm = np.linalg.norm(df['price'])
```

```python
df['price_norm'] = df['price'].apply(lambda x: x/norm)
```

```python
data = df[['sale_date', 'latitude', 'longitude', 'price_norm', 'price']]
```

```python
data
```

```python
data.shape
```

```python
gradient = {0.2: 'blue', 0.5: 'lime', 1: 'red'}
```

```python
data['price_norm'].min()
```

```python
data['price_norm'].max()
```

```python
#### remove missing locations
data.dropna(subset = ['longitude', 'latitude'], inplace = True)
```

```python
m = f.Map(location=[53.305494, -7.737649], tiles='Stamen Toner',  zoom_start=6, overlay = False)
```

```python
##data (list of points of the form [lat, lng] or [lat, lng, weight]) â€“ The points you want to plot. You can also provide a numpy.array of shape (n,2) or (n,3).
```

```python
hm = HeatMap(data = data[['latitude','longitude','price_norm']]).add_to(m)
```

```python
m.save(os.path.join('results', 'HeatMap.html'))
```

## Heat Map with time


```python
#format dates 
df_copy = data.copy()
```

```python
df_copy['sale_date'] = df_copy['sale_date'].apply(lambda x : datetime.datetime.strptime(x,'%d/%m/%Y'))
```

```python
df_copy.head()
```

```python
df_copy['sale_date'].sort_values().unique()
```

```python
df_hour_list = []
index = []
for sale_date in df_copy['sale_date'].sort_values().unique():
    df_hour_list.append(df_copy.loc[df_copy['sale_date'] == sale_date, ['latitude', 'longitude', 'price_norm']].groupby(['latitude', 'longitude']).sum().reset_index().values.tolist())
    index.append(sale_date)

```

```python
base_map = f.Map(location=[53.305494, -7.737649], tiles='Stamen Toner',  zoom_start=6, overlay = False)
HeatMapWithTime(df_hour_list, radius=25, gradient={0.2: 'blue', 0.4: 'lime', 0.7: 'orange', 1: 'red'}, min_opacity=0.5, max_opacity=0.8, use_local_extrema=True).add_to(base_map)
base_map.save(os.path.join('results', 'HeatMapWithTime.html'))
```

### try with data grouped by months

```python
#format dates 
df_copy['sale_month'] = df_copy['sale_date'].apply(lambda x : datetime.datetime(x.year, x.month, 1))
```

```python
df_copy.
```

```python
df_months= []
index = []
for sale_date in df_copy['sale_month'].sort_values().unique():
    df_months.append(df_copy.loc[df_copy['sale_month'] == sale_date, ['latitude', 'longitude', 'price']].groupby(['latitude', 'longitude']).sum().reset_index().values.tolist())
    index.append(datetime.datetime.utcfromtimestamp(sale_date.astype('O')/1e9).strftime("%m/%Y"))
```
